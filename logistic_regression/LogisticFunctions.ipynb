```python
import numpy as np
import matplotlib.pyplot as plt
```

Logit function

```python
x = np.linspace(-5, 5, 1000)
y = np.log(x / (1 - x))

plt.plot(x, y, lw=2)
plt.grid()
plt.savefig('figures/logit.pdf')
```

Sigmoid function

```python
x = np.linspace(-5, 5, 1000)
y = 1 / (1 + np.exp(-x))

plt.plot(x, y, lw=2)
plt.grid()
plt.savefig('figures/sigmoid.pdf')
```

Log loss

```python
x = np.linspace(0, 1, 1000)
y = -np.log(x)

plt.plot(x, y, lw=2)
plt.xlabel('$p$')
plt.ylabel('Log loss')
plt.grid()

plt.savefig('figures/log_loss.pdf')
```
