```python
import numpy as np
import matplotlib.pyplot as plt
```

Sigmoid function

```python
x = np.linspace(-5, 5, 1000)
y = 1 / (1 + np.exp(-x))

plt.plot(x, y, lw=2)
plt.grid()
plt.savefig('figures/sigmoid.pdf')
```

Derivative of sigmoid

```python
x = np.linspace(-5, 5, 1000)
sig = 1 / (1 + np.exp(-x))
y = sig * (1 - sig)

plt.plot(x, y, lw=2)
plt.grid()
```

Second derivative of sigmoid

```python
x = np.linspace(-5, 5, 1000)
sig = 1 / (1 + np.exp(-x))
y = sig**2 * (1 - sig) * (1 - 2 * sig)

plt.plot(x, y, lw=2)
plt.grid()
```

Tanh function

```python
x = np.linspace(-5, 5, 1000)
y = np.tanh(x)

plt.plot(x, y, lw=2)
plt.grid()
plt.savefig('figures/tanh.pdf')
```

Derivatives of tanh

```python
x = np.linspace(-5, 5, 1000)
y = 1 - np.tanh(x)**2

plt.plot(x, y, lw=2)
plt.grid()
```

Second derivative of tanh

```python
x = np.linspace(-5, 5, 1000)
y = -2 * np.tanh(x) * (1 - np.tanh(x)**2)

plt.plot(x, y, lw=2)
plt.grid()
```

ReLU function

```python
x = np.linspace(-5, 5, 1000)
y = np.where(x >= 0, x, 0)

plt.plot(x, y, lw=2)
plt.grid()
plt.savefig('figures/relu.pdf')
```

Leaky ReLU

```python
x = np.linspace(-5, 5, 1000)
alpha = 0.05
y = np.where(x >= 0, x, alpha * x)

plt.plot(x, y, lw=2)
plt.grid()
plt.savefig('figures/leaky_relu.pdf')
```
