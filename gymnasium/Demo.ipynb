```python
#env = gym.make('Taxi-v3', render_mode='human')
env = gym.make('Taxi-v3', render_mode='ansi')
```

```python
env.reset()
```

```python
print(env.render())
```

```python
for i in range(25):
    action = env.action_space.sample()
    print(env.render())
    observation, reward, terminated, truncated, _ = env.step(action)
    
env.close()
```
