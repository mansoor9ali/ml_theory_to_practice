```python
# Author: Roi Yehoshua <roiyeho@gmail.com>
# License: MIT

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(0)
```

```python
from sklearn.datasets import make_blobs

X, y = make_blobs(n_samples=500, cluster_std=[1.0, 2.5, 0.4], random_state=30)
```

```python
def plot_data(X):
    sns.scatterplot(x=X[:, 0], y=X[:, 1], edgecolor='k', legend=False)
    plt.xlabel('$x_1$')
    plt.ylabel('$x_2$')
```

```python
plot_data(X)
```

```python
from sklearn.mixture import GaussianMixture

gmm = GaussianMixture(n_components=3)
labels = gmm.fit_predict(X)
```

```python
def plot_clusters(X, labels):
    sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=labels, palette='tab10', edgecolor='k', legend=False)
    plt.xlabel('$x_1$')
    plt.ylabel('$x_2$')
```

```python
plot_clusters(X, labels)
```

```python
from sklearn.cluster import KMeans

kemans = KMeans(n_clusters=3, random_state=0) 
kemans.fit(X)
plot_clusters(X, kemans.labels_)
```

```python
from sklearn.cluster import AgglomerativeClustering

agg = AgglomerativeClustering(n_clusters=3, linkage='single')
agg.fit(X)
plot_clusters(X, agg.labels_)
```

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps=0.4)
dbscan.fit(X)
plot_clusters(X, dbscan.labels_)
```
