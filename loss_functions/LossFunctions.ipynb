Regression

```python
x = np.linspace(-4, 4, 1000)  # represents (y - \hat{y})

squared_loss = x**2
abs_loss = np.abs(x)
delta = 1
huber_loss = np.where(np.abs(x) <= delta, 0.5 * x**2, delta * np.abs(x) - 0.5 * delta**2)

plt.plot(x, squared_loss, c='b', lw=2, label='Squared loss')
plt.plot(x, abs_loss, c='r', lw=2, label='Absolute loss')
plt.plot(x, huber_loss, c='g', lw=2, label='Huber loss ($\delta = 1$)')

plt.xlabel('$y - \hat{y}$')
plt.ylabel('Loss')
plt.legend()

plt.savefig('figures/regression_loss_functions.pdf')
```

Binary Classification

```python
x = np.linspace(-4, 4, 1000)  # represents y * \hat{y}

log_loss = np.log(1 + np.exp(-x)) / np.log(2)
hinge_loss = np.where(1 - x >= 0, 1 - x, 0)
zero_one_loss = np.where(x >= 0, 0, 1)

plt.plot(x, log_loss, c='b', lw=2, label='Log loss')
plt.plot(x, hinge_loss, c='r', lw=2, label='Hinge loss')
plt.plot(x, zero_one_loss, c='k', lw=2, label='0-1 loss')

#plt.grid()
plt.xlabel('$y\hat{y}$')
plt.ylabel('Loss')
plt.legend()

plt.savefig('figures/binary_loss_functions.pdf')
```
