```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
from scipy.special import comb

def ensemble_error(n_classifiers, error):
    init_k = int(math.ceil(n_classifiers / 2))
    probs = [comb(n_classifiers, k) * error ** k * (1 - error) ** (n_classifiers - k)
             for k in range(init_k, n_classifiers + 1)]
    return sum(probs)
```

```python
error_range = np.arange(0.0, 1.0, 0.01)
errors = [ensemble_error(n_classifier=25, error=error) for error in error_range]

plt.plot(error_range, errors, label='Ensemble error', lw=2)
plt.plot(error_range, error_range, label='Base error', ls='--', lw=2)
plt.xlabel('Base error')
plt.ylabel('Ensemble error')
plt.legend()
plt.grid(alpha=0.5)

plt.savefig('figures/ensemble_error.pdf')
```
