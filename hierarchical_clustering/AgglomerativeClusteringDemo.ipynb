```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(0)
```

```python
from sklearn.datasets import make_blobs

X, y = make_blobs(n_samples=500, centers=3, cluster_std=0.6, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler

X = StandardScaler().fit_transform(X)
```

```python
def plot_data(X):
    sns.scatterplot(x=X[:, 0], y=X[:, 1], edgecolor='k', legend=False)
    plt.xlabel('$x_1$')
    plt.ylabel('$x_2$')
```

```python
plot_data(X)
plt.savefig('figures/blobs_data.pdf')
```

```python
from sklearn.cluster import AgglomerativeClustering

agg = AgglomerativeClustering(n_clusters=3, linkage='single')
agg.fit(X)
```

```python
def plot_clusters(X, labels):
    sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=labels, palette='tab10', edgecolor='k', legend=False)
    plt.xlabel('$x_1$')
    plt.ylabel('$x_2$')
```

```python
plot_clusters(X, agg.labels_)
plt.savefig('figures/blobs_single_linkage.pdf')
```

```python
agg = AgglomerativeClustering(n_clusters=3, linkage='complete')
agg.fit(X)

plot_clusters(X, agg.labels_)
plt.savefig('figures/blobs_complete_linkage.pdf')
```

```python
agg = AgglomerativeClustering(n_clusters=3, linkage='average')
agg.fit(X)

plot_clusters(X, agg.labels_)
plt.savefig('figures/blobs_average_linkage.pdf')
```

```python
agg = AgglomerativeClustering(n_clusters=3, linkage='ward')
agg.fit(X)

plot_clusters(X, agg.labels_)
plt.savefig('figures/blobs_ward_method.pdf')
```
